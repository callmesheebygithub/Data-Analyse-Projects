## TOTAL UNIQUE JOB

```python
job=data.pivot_table(columns=['Job'],aggfunc='size')
len(job)
```

```python
list1=[]
for item in data["Job"]:
    list1.append(item.split(','))
list1
```

```python
one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
one_d
```

```python
uniq_job=[]
for item1 in one_d:
    if item1 not in uniq_job:
        uniq_job.append(item1)
uniq_job
```

```python
len(uniq_job)
```

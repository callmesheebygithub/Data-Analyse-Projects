##  Find unique Values from Genre

```python
data['Genre']
```

```python
list1=[]
for value in data['Genre']:
    list1.append(value.split(','))
list1
```

```python
# 2D list convert into 1D list
oneDlist=[]
for item in list1:
    for item1 in item:
        oneDlist.append(item1)
oneDlist
```

```python
unique_list=[]
for item in oneDlist:
    if item not in unique_list:
        unique_list.append(item)
unique_list   
```
